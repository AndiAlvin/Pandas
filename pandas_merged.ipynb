```python

# import pandas
import pandas as pd
import numpy as np
  
# read csv data
df1 = pd.read_csv('Karyawan.csv')
df2 = pd.read_csv('Latihan.csv')
  
print(df1)
print(df2)
```

```python
inner_join = pd.merge(df1, 
                      df2, 
                      on ='Nama', 
                      how ='inner')
inner_join
```

```python
Left_join = pd.merge(df1, 
                     df2, 
                     on ='Nama', 
                     how ='left')
Left_join
```

```python
Right_join = pd.merge(df1, 
                      df2, 
                      on ='Nama',
                      how ='right')
Right_join
```

```python
Outer_join = pd.merge(df1, 
                      df2, 
                      on ='Nama', 
                      how ='outer')
Outer_join
```
